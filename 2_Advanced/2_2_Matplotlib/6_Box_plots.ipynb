```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_DS_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Scientist')].copy()
df_DS_US = df_DS_US.dropna(subset='salary_hour_avg')['salary_hour_avg']

#Plotting
df_DS_US.plot(kind='box', vert=False)
plt.ylabel('Data Scientist')
plt.xlabel('Hourly salary')

#Formatting tickers
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${x:.0f}'))
ax.set_yticklabels([])

plt.show()
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
df_DA.dropna(subset='salary_hour_avg', inplace=True)

countries = ['United States', 'Canada']
jobs_list = [df_DA[df_DA['job_country'] == country]['salary_hour_avg'] for country in countries]

plt.boxplot(jobs_list, tick_labels=countries, vert=False)
plt.xlim(0, 150)
plt.xlabel('Hourly salary')

#Formatting tickers
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${x:.0f}'))

plt.show()
```

```python
df_CA = df[df['job_country'] == 'Canada'].copy()
df_CA.dropna(subset='salary_hour_avg', inplace=True)

job_roles = ['Data Analyst', 'Data Scientist', 'Data Engineer']
jobs_list = [df_CA[df_CA['job_title_short'] == role]['salary_hour_avg'] for role in job_roles]

plt.boxplot(jobs_list, tick_labels=job_roles, vert=False)
plt.title('Comparison of Hourly Salaries for Data Roles in Canada')
plt.xlabel('Hourly salary')

#Formatting tickers
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${x:.0f}'))

plt.show()
```
