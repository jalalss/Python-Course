```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst']
plt.pie(df_DA['job_health_insurance'].value_counts(), startangle=90, labels=['No', 'Yes'], autopct='%1.1f%%')
plt.title('Proportion of Data Analyst Jobs w/ Health Insurance')
plt.show()
```

```python
df_ft_pt = df[(df['job_schedule_type'] == 'Full-time') | (df['job_schedule_type'] == 'Part-time')]
plt.pie(df_ft_pt['job_schedule_type'].value_counts(), startangle=90, autopct='%1.1f%%', labels=['Full-time', 'Part-time'])
plt.title('Full-time vs. Part-time job postings')
plt.show()
```

```python
df_US = df[df['job_country'] == 'United States']
plt.pie(df_US['job_no_degree_mention'].value_counts(), startangle=90, labels=['Degree', 'No degree'], autopct='%1.1f%%', explode=(0, 0.05))
plt.title('Proportion of US Jobs w/ No Degree Mentions')
plt.show()
```
