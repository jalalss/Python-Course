```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
import calendar

## Get Cloud Engineer jobs only
df_ce = df[df['job_title_short'] == 'Cloud Engineer'].copy()

## Create a month column
df_ce['job_posted_mo'] = df_ce['job_posted_date'].dt.to_period('M')

cloud_eng_job_trend = df_ce.groupby('job_posted_mo').size()

## Plotting
cloud_eng_job_trend.plot(
    kind='line',
    linestyle = ':',
    color = 'purple',
    marker = 'o'
)

plt.title('Job Postings Trend for Cloud Engineers')
plt.xlabel('')
plt.ylabel('Job postings count')
plt.show()
```

```python
df_sda = df[df['job_title_short'] == 'Senior Data Analyst'].copy()
top_companies_sda = df_sda.groupby('company_name').size().sort_values(ascending=False).head(5)

colors = ['blue', 'orange', 'green', 'red', 'purple']

top_companies_sda.plot(
    kind = 'bar',
    color = colors,

)
plt.grid(axis = 'y', linestyle = '--')
plt.title('Top 5 Companies hiring Senior Data Analysts by Number of Postings', fontsize='14')
plt.ylabel('Number of job postings')
plt.xlabel('')
plt.xticks(rotation = 45, ha='right')
plt.show()
```
