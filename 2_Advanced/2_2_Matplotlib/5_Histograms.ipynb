```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df_DS = df[df['job_title_short'] == 'Data Scientist'].copy()
df_DS['salary_hour_avg'].plot(kind='hist', bins=40, color='blue', edgecolor='black', linewidth=1)
plt.xlabel('Hourly salary')
plt.ylabel('Number of job posting')
plt.title('Distribution of Data Scientist hourly salaries')

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}$'))
plt.show()
```

```python
df_skills = df.dropna(subset=['job_skills']).copy()
df_skills['num_skills'] = df_skills['job_skills'].apply(len)

df_skills['num_skills'].plot(kind='hist', bins=55, color='magenta', edgecolor='black')
plt.xlabel('Number of skills')
plt.ylabel('Number of job postings')
plt.title('Distribution of job postings by number of skills required')
plt.show()
```

```python
df_DE = df[df['job_title_short'] == 'Data Engineer']
df_DE['salary_year_avg'].plot(kind='hist', bins=50, color='red', edgecolor='black')

#Formatting x-tickers
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}k'))

#Adding labels
plt.title('Salary distribution of Data Engineering job postings')
plt.xlabel('Yearly salary')
plt.ylabel('Number of job postings')

#Adding annotations
mean = df_DE['salary_year_avg'].mean()
median = df_DE['salary_year_avg'].median()
plt.axvline(mean, color='blue', linestyle='dashed', label=f'Mean: ${mean:,.0f}')
plt.axvline(median, color='green', linestyle='dashed', label=f'Median: ${median:,.0f}')
plt.legend()
plt.show()
```
