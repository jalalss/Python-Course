```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda job_posted_dt : job_posted_dt.strftime('%Y-%m-%d'))
df[['job_posted_date', 'job_posted_date_str']]
```

```python
from datetime import datetime

df['days_since_posted'] = df['job_posted_date'].apply(lambda job_posted_dt: (datetime.now() - job_posted_dt).days)
df[['days_since_posted', 'job_posted_date']].head()
```

```python
df_filtered = df.copy().dropna(subset='salary_year_avg').reset_index()

df_filtered['salary_category'] = df_filtered['salary_year_avg'].apply(lambda salary : 'Low' if salary < 60_000 else 'Medium' if (salary >= 60_000 and salary <= 100_000) else 'High')
df_filtered[['salary_category', 'salary_year_avg']]
```
