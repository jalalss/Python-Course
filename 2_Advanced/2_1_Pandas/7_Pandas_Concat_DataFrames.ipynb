```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df
```

```python
data_scientist_df = df[df['job_title_short'] == 'Data Scientist']
data_analyst_df = df[df['job_title_short'] == 'Data Analyst']

combined_df = pd.concat([data_analyst_df, data_scientist_df], ignore_index=True)
combined_df
```

```python
## Create month column
df_copy = df.copy()
df_copy['job_posted_month'] = df_copy['job_posted_date'].dt.strftime('%b')

## Create Q4 data frames
oct_df = df_copy[df_copy['job_posted_month'] == 'Oct']
nov_df = df_copy[df_copy['job_posted_month'] == 'Nov']
dec_df = df_copy[df_copy['job_posted_month'] == 'Dec']

df_q4 = pd.concat([oct_df, nov_df, dec_df], ignore_index=True)
df_q4['job_posted_month'].value_counts().plot(kind='bar')
```

```python
ft_df = df[df['job_schedule_type'].str.contains('Full-time', na=False)]
pt_df = df[df['job_schedule_type'].str.contains('Part-time', na=False)]
combined_schedules_df = pd.concat([ft_df, pt_df], ignore_index=True).drop_duplicates(subset=['job_title', 'company_name', 'job_posted_date'])
combined_schedules_df
```
