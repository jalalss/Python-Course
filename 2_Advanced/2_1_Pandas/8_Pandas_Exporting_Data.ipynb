```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.to_csv('data_fame.csv', index=False)
pd.read_csv('data_fame.csv')
```

```python
df_data_scientist = df[df['job_title_short'] == 'Data Scientist']
df_data_scientist.to_csv('df_data_scientist.csv', index=False)
pd.read_csv('df_data_scientist.csv')
```

```python
nyc_jobs = df[df['job_location'] == 'New York, NY']
nyc_jobs.to_excel('nyc_jobs.xlsx', index=False)
```
