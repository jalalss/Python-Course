## Excercise

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import calendar

## Add month column
df['job_posted_month'] = df['job_posted_date'].dt.month

## Remove some job titles
job_titles = ['Data Scientist', 'Data Analyst', 'Data Engineer']
df = df.loc[df['job_title_short'].isin(job_titles)]

## Add a month column
df['job_posted_month_name'] = df['job_posted_date'].dt.month

## Create pivot table
job_title_salary_trend = df.pivot_table(columns='job_title_short', index='job_posted_month', values='salary_year_avg', aggfunc='median')

# Convert numeric months (1..12) to month names.
job_title_salary_trend.index = job_title_salary_trend.index.map(lambda m: calendar.month_name[m])

# Plot each job title as a separate line
plt.plot(job_title_salary_trend)

# Label axes and set title
plt.xticks(rotation=45, ha='right')
plt.xlabel("Month")
plt.ylabel("Median Yearly Salary")
plt.title("Median Salary Trend by Job Title")

# Show legend with job titles (the columns of the transposed DataFrame)
plt.legend(job_title_salary_trend.columns, title="Job Titles")

# Display the plot
plt.show()
```

# 6_Pandas_Merge_DataFrames

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month

## US jobs only
df_us_jobs = df[df['job_country'] == 'United States']
```

```python

df_monthly_postings = df.groupby('job_posted_month').size().reset_index(name='postings_count')
df_monthly_postings
```

```python
df_us_jobs.merge(df_monthly_postings, on='job_posted_month', how='right').head()
```

```python
df_not_remote_jobs = df[df['job_work_from_home'] == False]
```

```python
df_not_remote_salary = df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].agg('mean').reset_index(name='salary_not_remove_avg')
df_not_remote_salary
```

```python
df_not_remote_jobs.merge(df_not_remote_salary, on='job_title_short', how='right').head()
```

```python
df.drop(columns='job_posted_month', inplace=True)
```

```python
df_copy = df.copy()
##df_copy.dropna(subset='salary_year_avg', inplace=True)
df_company_salary = df_copy.groupby('company_name')['salary_year_avg'].agg('mean').reset_index(name='salary_year_avg')
df_company_salary
```

```python
df_company_remote = df[df['job_work_from_home'] == True].groupby('company_name').size().reset_index(name='remote_count')
df_company_remote
```

```python
df_company_salary.merge(df_company_remote, on='company_name', how='right').head()
```
