```python
class JobPosting:

    def __init__(self, title, company, location, salary = 0):
        self.title = title
        self.company = company
        self.location = location
        self.salary = salary

    def to_dict(self):
        return {
            'title' : self.title,
            'company' : self.company,
            'location' : self.location,
            'salary' : self.salary
        }

    def compare_salary(self, other_job):
        if self.salary > other_job.salary:
            return f'{self.title} pays more than {other_job.title}'
        elif self.salary == other_job.salary:
            return f'{self.title} and {other_job.title} pay the same'
        else:
            return f'{self.title} pays less than {other_job.title}'
```
